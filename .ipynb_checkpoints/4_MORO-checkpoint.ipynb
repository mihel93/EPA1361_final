### 4. Multi-Objective Robust Optimization (MORO)

In notebook "3_comparison", we identified that any policy will help in reducing the "expected number of deaths", the "expected annual damage" in both Provinces, as well as the ecacuation cost and Rfr cost. Because every policy will decrease these variables, it is our task to find the most favourable and acceptable policy for the involved actors, regarding their interests and preferences. And that is where Many Objective Robust Optimization (MORO) comes in. MORO is a form of robust decision making that takes robustness into account in the search phase directly (Bartholomew & Kwakkel, 2020), by optimizing over robustness metrics rather than over the KPI's themselves.

As can be seen in the report, the aim of this analysis is to find a policy measure for Overijssel province that find is both robust and accepted by the different political parties. By using MORO, the robustness aspect can at least be achieved. And by taking into account the variance over locations, and by taking into account the most vulnerable locations, the acceptability part is aimed to be achieved.

MORO optimizes over the robustness-values generated by the self-defined KPIs of a policy. These robustness-values are said to be minimized or maximized in order to find an optimal policy within the space we limit by constraints and our definition of robustness.

With the retrieved "outcomes_combined" dataframe, we can aggregate the variables in order to visualize them against each other.

We implement the aggregate function 'aggregate_kpi' created in notebook "1_open_exploration_base_case" and calculate the aggregated variables.

```python
# The following function is used to aggregate the outcomes into the KPI we want
# This iterates over all the locations and round numbers, and creates a new column summing the values per location and round.
# If we want to aggregate over the location, "aggregate" equals "location" and therefore the KPI is added per location and not in total.
# If we want to aggregate over the province, "aggregate" equals "name of the province" and therefore the KPI is added per location and not in total
# On the contrary, if aggregate equals "total", the total value is appended to the dataframe.
def aggregate_kpi(data, kpi, aggregate):
    locations = ["A.1", "A.2", "A.3", "A.4", "A.5"]
    kpi_columns = []
    
    if kpi == "RfR Total Costs" or kpi == "Expected Evacuation Costs":
        kpi_columns.append(kpi + " 0")
        kpi_columns.append(kpi + " 1")
        kpi_columns.append(kpi + " 2")
        
        data[kpi] = data[kpi_columns].sum(axis=1)
    else:
        if aggregate == "total":
            for location in locations:
                kpi_columns.append(location + "_" + kpi + " 0")
                kpi_columns.append(location + "_" + kpi + " 1")
                kpi_columns.append(location + "_" + kpi + " 2")

            data[kpi] = data[kpi_columns].sum(axis=1)
            
        elif aggregate == "Overijssel":
            for location in locations:
                if location == "A.4" or location == "A.5":
                    kpi_columns = []
                    kpi_columns.append(location + "_" + kpi + " 0")
                    kpi_columns.append(location + "_" + kpi + " 1")
                    kpi_columns.append(location + "_" + kpi + " 2")
                    
            data["Overijssel" + "_" + kpi] = data[kpi_columns].sum(axis=1)
      
        elif aggregate == "Gelderland":
            for location in locations:
                if location == "A.1" or location == "A.2"  or location == "A.3":
                    kpi_columns = []
                    kpi_columns.append(location + "_" + kpi + " 0")
                    kpi_columns.append(location + "_" + kpi + " 1")
                    kpi_columns.append(location + "_" + kpi + " 2")

            data["Gelderland" + "_" + kpi] = data[kpi_columns].sum(axis=1)
            
        else:
            for location in locations:
                kpi_columns = []
                kpi_columns.append(location + "_" + kpi + " 0")
                kpi_columns.append(location + "_" + kpi + " 1")
                kpi_columns.append(location + "_" + kpi + " 2")

                data[location + "_" + kpi] = data[kpi_columns].sum(axis=1)
                
    return data
```

### Robustness functions ###

In an ideal situation, policy would be effective in any scenario. With a robustness function we can calculate the robustness of a policy over a set of scenarios, i.e. how well they would perform in many different uncertain futures. In this case, signal-to-noise is chosen as a robustness metric, the reasoning for which can be found in chapter 2. As all the KPI's considered are to be minimized, a robust policy can be defined as one where the mean (=signal) is low, as well as the standard deviation (=noise). 

The optimization algorithm outputs an one-dimensional array of the variable, with an outcome for each scenario and each outcome. Let's say we want to calculate the robustness over 3 variables we would like to aggregate with 5 scenarios. The input will be an array with 3 arrays of 5 items. First we sum over the 3 arrays, resulting in a one-dimensional array with the 5 summed values. Then, we calculate the mean and the std of this array and multiply them with each other. This value is going to be minimized by the MORO-algoritm.

The signal-to-noise function adds 0.001 to the mean and the standard deviation. This is because values of the "Expected Number of Deaths" and "Evacuation Costs" can be zero. In this way, if the mean is a valid number and the standard deviation is zero (because all the values within the scenario are the same), we prevent the robustness values from getting the value zero. 

```python
#Function to calculate signal to noise value for the 'normal' KPI's, i.e. expected numer of casualties, expected annual damage and total non-RfR policy costs.

def signal_to_noise(*data):
        
    dataset = sum(data)  
    mean = np.mean(dataset)
    std = np.std(dataset)
    return (mean + 0.001) * (std + 0.001)

# Because for the variation between the locations KPI, we first need to calculate the one-dimensional array
# with the std between them, instead of the sum. This done per 6 variables of input.
# So per pair of 6 variables, the function calculates the standard deviation per column instead of per row (axis=0).
# The result is a one-dimensional array with the standard deviation between the variables. Within this one-dimensional array
# We calculate the signal-to-noise of these stdvs. 
# The calculation of the one-dimensional array with the std values is hardcoded and is assumed to have the form of 
# pairs of 6 to calculate the std over. 
# This function is only used in the MORO itself, since the variance between locations is not 
# an outcome of the problem formulation
def signal_to_noise_variation(*data):
    invest_costs_location = [sum(data[x:x+6]) for x in range(0, len(data), 6)]
    std_arr = np.array(invest_costs_location).std(axis=0)
    
    var_mean = np.mean(std_arr)
    var_std = np.std(std_arr)

    return (var_mean * var_std)
```

**Hypervolume indicator values**

In order to find optimal policies with sufficient certainty, the multi-objective evolutionary algorithm used by MORO needs to converge. An indicator for this is Hypervolume. This metric needs a value for the minimum and maximum expected robustness value for each KPI. Therefore, we apply the robustness functions on the random policy run executed in "3_comparison". In this way, we get a rough estimation of the minimum and maximum values we can expect from the MORO and to look for in the hypervolume space.

```python
# Select the KPIs and apply the robustness function to it
df_ooi = combined.loc[:,["policy",'Expected Number of Deaths','RfR Total Costs',
                            'Expected Evacuation Costs', 'Overijssel_Expected Annual Damage',
                            'Gelderland_Expected Annual Damage','Overijssel_Dike Investment Costs',
                            'Gelderland_Dike Investment Costs']]
robust_stn_values = df_ooi.groupby(by="policy").apply(signal_to_noise)

robust_stn_values.head()
```

```python
# We define the min and max values for the KPIs we want to optimize over

min_deaths = robust_stn_values["Expected Number of Deaths"].min()
max_deaths = robust_stn_values["Expected Number of Deaths"].max()

min_evac = robust_stn_values["Expected Evacuation Costs"].min()
max_evac = robust_stn_values["Expected Evacuation Costs"].max()

min_rfr = robust_stn_values["Gelderland_Dike Investment Costs"].min()
max_rfr = robust_stn_values["Gelderland_Dike Investment Costs"].max()

min_annual_Gelderland = robust_stn_values["Gelderland_Expected Annual Damage"].min()
max_annual_Gelderland = robust_stn_values["Gelderland_Expected Annual Damage"].max()

min_annual_Overijssel = robust_stn_values["Overijssel_Expected Annual Damage"].min()
max_annual_Overijssel = robust_stn_values["Overijssel_Expected Annual Damage"].max()

```

????Let's now define the constraints. Whereas previously (notebook '2 - Comparison'), total policy costs were aggregated, this forms a problem for MORO, as the RfR costs are the same per policy (there is no standard deviation when the whole array has the same value) and therefore the signal to noise value will always be zero. Therefore, we chose to add RfR costs as a constraint, and to consider the remaining policy costs (= Dike Investment Costs + Evacuation costs) separately, as they *do* have different values per policy. This simulataneously reflects many actors preferring dyke heightening over Room for the River, which will now be incorporated due to this constraint.

What value to constraint RfR costs at? We assume that RfR costs should not exceed the median value (= 50% of its values) that was found when performing the run with random policy.  In this way, costs are kept low.

**Defining the uncertainty space**

From "1_open_exploration_base_case" we concluded that the worst scenario's consisted of a certain range of uncertainty-values. Therefore, we redefine the range of these uncertainties accordingly. We assume that if the policy we will find is robust under the worst-case scenario's, it definitely will be robust under less worse circumstances. Also, this will speed up the optimization a bit, since the scenario space will be smaller and therefore, we can reduce our number of scenarios.

```python
# Instantiate the model
dike_model, planning_steps = get_model_for_problem_formulation(5)

def sum_over(*args):
    return sum(args)

# Redefine the uncertainties
dike_model.uncertainties['A.1_pfail'] = RealParameter('A.1_pfail', 0, 0.4)
dike_model.uncertainties['A.3_pfail'] = RealParameter('A.5_pfail', 0, 0.21)
dike_model.uncertainties['A.1_Bmax'] = RealParameter('A.1_Bmax', 30, 330)
```

### MORO Set up ###

As explained earlier, we use the signal-to-noise function (mean * std) to calculate the robustness values of a policy. This function will be used to minimize:

- The total expected number of deaths
- The total expected annual damage
- The total investment costs for dikes (later combined to 'Total non-RfR policy costs')
- The total evacuation costs (later combined to 'Total non RfR policy costs')
- The variance of costs between locations


Since the RfR costs are the same per scenario of a policy, there is no point in calculating the signal-to-noise of this variable. To solve that the optimization will not keep increasing the RfR costs, the aforementioned constraint is implemented. 

To calculate the total aggregated variables, all disaggregated variables are used as input. Only in this way, we can calculate the signal-to-noise of the aggregated variable.

```python
MINIMIZE = ScalarOutcome.MINIMIZE

#Making lists of all the disaggregated variables in order to properly use the total aggregated variables

deaths = ['A.1_Expected Number of Deaths 0', 'A.1_Expected Number of Deaths 1', 'A.1_Expected Number of Deaths 2',
                   'A.2_Expected Number of Deaths 0', 'A.2_Expected Number of Deaths 1', 'A.2_Expected Number of Deaths 2',
                   'A.3_Expected Number of Deaths 0', 'A.3_Expected Number of Deaths 1', 'A.3_Expected Number of Deaths 2',
                   'A.4_Expected Number of Deaths 0', 'A.4_Expected Number of Deaths 1', 'A.4_Expected Number of Deaths 2',
                   'A.5_Expected Number of Deaths 0', 'A.5_Expected Number of Deaths 1', 'A.5_Expected Number of Deaths 2']

rfr_costs = ['RfR Total Costs 0', 'RfR Total Costs 1', 'RfR Total Costs 2']

evac_cost = ['Expected Evacuation Costs 0','Expected Evacuation Costs 1', 'Expected Evacuation Costs 2']

dikes_Gelderland = ['A.1_Dike Investment Costs 0', 'A.1_Dike Investment Costs 1', 'A.1_Dike Investment Costs 2',
                 'A.2_Dike Investment Costs 0', 'A.2_Dike Investment Costs 1', 'A.2_Dike Investment Costs 2',
                 'A.3_Dike Investment Costs 0', 'A.3_Dike Investment Costs 1', 'A.3_Dike Investment Costs 2']

dikes_Overijssel = ['A.4_Dike Investment Costs 0', 'A.4_Dike Investment Costs 1', 'A.4_Dike Investment Costs 2',
                 'A.5_Dike Investment Costs 0', 'A.5_Dike Investment Costs 1', 'A.5_Dike Investment Costs 2']

annual_dmg_Gelderland = ['A.1_Expected Annual Damage 0', 'A.1_Expected Annual Damage 1', 'A.1_Expected Annual Damage 2',
                   'A.2_Expected Annual Damage 0', 'A.2_Expected Annual Damage 1', 'A.2_Expected Annual Damage 2',
                   'A.3_Expected Annual Damage 0', 'A.3_Expected Annual Damage 1', 'A.3_Expected Annual Damage 2']
                  
annual_dmg_Overijssel = ['A.4_Expected Annual Damage 0', 'A.4_Expected Annual Damage 1', 'A.4_Expected Annual Damage 2',
                   'A.5_Expected Annual Damage 0', 'A.5_Expected Annual Damage 1', 'A.5_Expected Annual Damage 2']


dikes_cost = ['A.1_Dike Investment Costs 0', 'A.1_Dike Investment Costs 1', 'A.1_Dike Investment Costs 2',
                 'A.2_Dike Investment Costs 0', 'A.2_Dike Investment Costs 1', 'A.2_Dike Investment Costs 2',
                 'A.3_Dike Investment Costs 0', 'A.3_Dike Investment Costs 1', 'A.3_Dike Investment Costs 2',
                 'A.4_Dike Investment Costs 0', 'A.4_Dike Investment Costs 1', 'A.4_Dike Investment Costs 2',
                 'A.5_Dike Investment Costs 0', 'A.5_Dike Investment Costs 1', 'A.5_Dike Investment Costs 2']



#Specifying robustness functions. All have to be minimized. All except for the variance will use the 'signal_to_noise' function.
#Variance will use the 'signal_to_noise_variation' function, which contains the extra step of first calculating the standard deviation.
robustness_functions = [ScalarOutcome('Death Score', kind=MINIMIZE, variable_name=deaths, function=signal_to_noise),
                        ScalarOutcome('Annual Costs Gelderland Score', kind=MINIMIZE, variable_name=annual_dmg_Gelderland,
                                      function=signal_to_noise),
                        ScalarOutcome('Annual Costs Overijssel Score', kind=MINIMIZE, variable_name=annual_dmg_Overijssel,
                                      function=signal_to_noise),
                        ScalarOutcome('RfR Cost Score', kind=MINIMIZE, variable_name=rfr_costs, function=signal_to_noise),
                        ScalarOutcome('Evacuation Score', kind=MINIMIZE, variable_name=evac_cost, function=signal_to_noise)]

#Adding the constraint for Room for the River costs.
constraints = [Constraint("dike_cost", outcome_names=dikes_Gelderland,
                          function=lambda x:max(0, x-100000))]
```

We made the uncertainty space smaller by implementing the ranges in which the worst-case scenarios occurred. Therefore, we made a decision to use a number of scenarios of 20. Since all of the scenarios are in the worst-case space, running 100 scenarios will only delay the run.

For the epsilon values, high values have been chosen. Since the signal-to-noise robustness function produces very large numbers and costs is a large number of itself, the epsilon values do not have to be very small. The exact values have been iteratively determined, by on one hand performing small runs and seeing the amount of solutions that came out of it. On the other hand, the table above that shows the robustness values for the 'random policy' run gives a rough insight in the produced values. Based on these values, epsilon values have been chosen that are not too small to not make any difference, but not too large to not be able to trade off between policies that have significantly different results. By making the epsilon values quite large, the run is also sped up.

```python
#Performing MORO: specifying the scenarios, the convergence metrics, the number of function evaluations and performing the robust_optimize function.
n_scenarios = 20
scenarios = sample_uncertainties(dike_model, n_scenarios)

convergence = [HyperVolume(minimum=[min_deaths, min_evac, min_rfr, min_annual_Gelderland, min_annual_Overijssel], 
                           maximum=[max_deaths, max_evac, max_rfr, max_annual_Gelderland, max_annual_Overijssel]),
               EpsilonProgress()]

nfe = int(100)

with MultiprocessingEvaluator(dike_model) as evaluator:
    archive, convergence = evaluator.robust_optimize(robustness_functions, scenarios, 
                                               nfe=nfe, convergence=convergence,
                                               epsilons=[0.01,]*len(dike_model.outcomes),
                                                    constraint=constraints)
with open('results/MORORun2.pkl', 'wb') as file_pi:
    pickle.dump((archive, convergence), file_pi)
```

**Convergence**

In order to track whether the MORO has converged (i.e., no better solutions can be found within the demarcated policy space) to the possible optimum solutions, we analyse the convergence metrics. This will be done by analysing the epsilon-convergence and the hypervolume-convergence.

By looking at both the epsilon-progress and the hypervolume, we can conclude that both of them are convergenced. However, for the hypervolume, we may have picked ranges that were a bit too small (since the max value of 1 is reached). However, we assume that the epsilon-progress is the most important one and that the small ranges of hypervolume are not that big of a problem. Therefore, we can conclude the algorithm converged.  

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharex=True, figsize=(8, 4))
ax1.plot(convergence.nfe, convergence.epsilon_progress)
ax1.set_ylabel('$\epsilon$-progress')
ax2.plot(convergence.nfe, convergence.hypervolume)
ax2.set_ylabel('hypervolume')

ax1.set_xlabel('number of function evaluations')
ax2.set_xlabel('number of function evaluations')
plt.show()
```
